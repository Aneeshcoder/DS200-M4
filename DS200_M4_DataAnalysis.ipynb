##Import Data
Import data from Github and display the DataFrame to see the dataset.



```python
import pandas as pd
from io import StringIO
import requests

csv_url = "https://raw.githubusercontent.com/Aneeshcoder/DS200-M4/refs/heads/main/MGNREGA.csv"
response = requests.get(csv_url)
response.raise_for_status()
df = pd.read_csv(StringIO(response.text))

display(df.head())
```

## Dataset Description
Displaying the column names, data types, and shape of the DataFrame for understanding the dataset.



```python
print("Column Names:")
display(df.columns)

print("\nData Types:")
display(df.dtypes)

print("\nShape of the DataFrame:")
display(df.shape)
```

## Dataset Plotting and Analysis
Display different relevant plots for understanding the dataset and analysis.



```python
import matplotlib.pyplot as plt
import seaborn as sns
```

''Persondays_of_Central_Liability_so_far'' represents the cumulative number of persondays for which the central government is liable for wage payments under the MGNREGA scheme up to a certain point in time.

```python
grouped = df.groupby('fin_year')['Persondays_of_Central_Liability_so_far']
mean = grouped.mean()
std = grouped.std()

plt.figure(figsize=(12,6))
plt.plot(mean.index, mean.values, label='Mean Persondays')
plt.fill_between(mean.index, mean-std, mean+std, alpha=0.25, label='Std. Dev.')
plt.xlabel('Year')
plt.ylabel('Persondays Generated')
plt.legend()
plt.title('Persondays Generated Over Years')
plt.show()
```

This plot shows the mean and standard deviation of persondays generated each financial year, indicating the overall trend. This plot shows the general trend that high Persondays generated generally equivalent to high data variation.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Total_Exp', y='Persondays_of_Central_Liability_so_far', hue='fin_year')
plt.legend(title='Financial Year')
plt.ylabel('Persondays Generated')
plt.xlabel('Total Expenditure')
plt.title('Funds Released v/s Persondays Generated')
plt.show()
```

This scatter plot visualizes the relationship between the total expenditure and the persondays generated. This plot helps to see the almost linear correlation between the funds spent and the amount of employment generated.

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='district_name', y='Persondays_of_Central_Liability_so_far', data=df)
plt.xticks(rotation=90)
plt.ylabel('Persondays Generated')

plt.title('Persondays Generated by District')
plt.show()
```

This box plot displays the distribution of persondays generated for each district. This plot shows that Hisar district generated maximum Persondays while the Faridabad district generated the minimum.

```python
plt.figure(figsize=(32,24))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='inferno')
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
plt.title('Correlation Matrix', fontsize = 18)
plt.show()
```

```python
average_wage_by_year = df.groupby('fin_year')['Average_Wage_rate_per_day_per_person'].mean()
plt.figure(figsize=(12, 6))
plt.plot(average_wage_by_year.index, average_wage_by_year.values, marker='o')
plt.xlabel('Financial Year')
plt.ylabel('Average Wage Rate per Day per Person')
plt.title('Trend of Average Wage Rate Over Years')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

This line plot shows the average wage rate per day per person for each financial year. This plot shows an increase in the average wage rate over the years with a sudden surge from 2024-25 to 2025-26.

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Average_days_of_employment_provided_per_Household', kde=True)
plt.title('Distribution of Average Days of Employment Provided per Household')
plt.xlabel('Average Days of Employment')
plt.ylabel('Frequency')
plt.show()
```

This histogram shows the frequency distribution of the Average days of employment provided per Household column. The x-axis represents the average number of days of employment, and the y-axis shows how many data entries fall within each range of days. The KDE (Kernel Density Estimate) line provides a smoothed representation of the distribution. This plot shows the average days of employement provided per household lies around 20 days.
